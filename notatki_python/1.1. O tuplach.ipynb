<h4> O tuplach / ( pl. krotkach ) </43>

```python
tuple1 = (1, 2.0, "3", ( 4,5,6 ), [1,3], {1,2,3,4,5}) # Tuple mogą być zagnieżdżane / wielowymiarowe; oraz mogą zawierać różne struktury i zmienne różnego typu.
print( f"tuple(tuple1) = {type(tuple1)}")  # Print the type of the tuple you created
print( f"tuple1[3][2] = {tuple1[3][-1]}") # -1 to ostatni element listy
print( f"tuple1[3][2] = {tuple1[2:4]}") # 2:4; elementy od index[2] do indexu 3 włącznie, bez indexu 4.

tuple2 = tuple1 + ( 8, 9 , 10 ) # tuple można konkatenować za pomocą znaku +
print( f"tuple2 = {tuple2}, lenght = { len(tuple2) }")
```
