```python
Lista1 = [ 1, (2,3), "4", 5.0 ]

print(f"Lista 1 = {Lista1}")
Lista1.append([6,7,8]) # append dodaje jeden element do listy; jedną zmienną, lub jedną strukturę.
print(f"Lista1 po .append([6,7,8]) = {Lista1}")
Lista1.extend([9,10,11]) # extend konkatenuje listę, dodając wszystkie jej elementy
print(f"Lista1 po .extend([9,10,11]) = {Lista1}")
Lista1 += [ 12,13,14 ] # Działa jak extend
print(f"Lista1 po + [12,13,14] = {Lista1}")

Lista1[0:len(Lista1)-2] = [ 10 ] # Takie operacje pozwalają na na poszerzanie i pomniejszanie list, wraz ze zmianą wartości
print(f"Lista1 po Lista1[0:3] = [ 10, 10 ] => {Lista1}")
Lista1[0:2] = [ "x", "y", "z" ]
print(f"Lista1 po Lista1[0:2] = [ 'x','y','z' ] => {Lista1}")
Lista1[0:3] = [] # Usunięcie elementów o danym indeksie
print(f"Lista1 po Lista1[0:4] = [] => {Lista1}")


lista2 = [1,2,3]
lista3 = lista2 # Najwyraźniej takie kopiowanie wartości powinno się robić za pomocą lista2[:], ale to też kopiuje wartość, jak widać poniżej
lista3 = [0]

print(lista2,lista3)

shopping_list = ['Watch', 'Laptop', 'Shoes', 'Clothes', 'Football'] #Zamiana elementu listy
shopping_list[ shopping_list.index("Laptop") ] = "Notebook"
```
