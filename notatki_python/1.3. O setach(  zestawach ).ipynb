```python
# set / zestaw; Zawiera unikalne wartości; automatycznie usuwa duplikaty.

set1 = {"pop", "rock", "soul", "hard rock", "rock", "R&B", "rock", "disco"}
print(f"set1 = {set1}")
# sety używają add i remove zamiast indeksów.
set1.add("test"); set1.remove("pop")
print(f"set1 po add('test') i remove('pop') => {set1}")

# Pokrycia setów / operacje na zestawch
set1 = {1,2,3,4}; set2 = {4,5,6,7}; set3 = set1 & set2 #pokrycie się 2 zestawów
print(f"set3 = set1 & set2 => {set3}")

intersection = set1.intersection(set2) # część wspólna zestawów, tak jak z operatorem &
difference = set1.difference(set2) #różnica zestawów w secie1; czyli odjęcie set2 od set1
union = set1.union(set2) # połączenie zestawów.


set1 = {1,2,3,4}; set2 = {1,3}
is_superset = set1.issuperset(set2) # czy set1 jest "supersetem" set2; czy set1 mieści się całkowicie w set2
print( is_superset )
is_subset = set1.issubset(set2) # czy set1 częścią set2
print( is_subset )
```
