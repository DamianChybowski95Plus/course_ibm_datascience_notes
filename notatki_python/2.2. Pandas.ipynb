DataFrame, czyli tabela, excelowa; z podobnymi metodami.

```python
# let us import the Pandas Library
import pandas as pd
```

```python
dataFrame = pd.DataFrame({ #Deklaracja tabeli za pomocą słownika
    'Name': ['Rose','John', 'Jane', 'Mary'], 
    'ID': [1, 2, 3, 4], 'Department': ['Architect Group', 'Software Group', 'Design Team', 'Infrastructure'], 
    'Salary':[100000, 80000, 50000, 60000]
})
display(dataFrame)
```

```python
#Retrieving the "ID" column and assigning it to a variable x
columnId = dataFrame[['ID']]
display(columnId, type(columnId))
```

```python
#write your code here
df_marks = pd.DataFrame({
    "Student" : [ "David", "Samuel", "Terry", "Evan" ],
    "Age" : [ 27, 24, 22, 32 ],
    "Country" : [ "UK", "Canada", "China", "USA" ],
    "Course" : [ "Python", "Data Structures", "Machine Learning", "Web Developement" ],
    "Marks" : [ 85, 72, 89, 76 ]
})

display(df_marks)
```

```python
# Odczytywanie kolumn
b = df_marks["Marks"]
c = df_marks[["Country", "Course"]]

# Odczytywanie za pomocą indexu; iloc() # kwadratowe nawiasy, mimo że to metoda.

display( df_marks.iloc[0,0] ) # lepiej pisać w jednym kwadratowym nawiasie kolejne argumenty, ponieważ to umożliwia dodatkową funkcjonalność;
print( df_marks.iloc[0,0:len(df_marks)] ) #
display( df_marks.iloc[0:4,2:4] ) # 
display( df_marks ) #

# Odczytywanie za pomocą labels; loc() - w przypadku gdy index jest domyślny, wiersze będą odczytywane za pomocą również liczb, czy dat
display( df_marks.loc[2:4,("Student","Age")]) # wiersze 2 i 3, kolumny student i age
```
